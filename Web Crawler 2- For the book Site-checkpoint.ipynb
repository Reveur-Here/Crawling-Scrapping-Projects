```python
import requests
from csv import writer
from csv import reader
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import webbrowser
```

```python
url="https://b-ok.cc/categories"
uClient = uReq(url)
page_html = uClient.read()
uClient.close()
```

```python
soup1=soup(page_html, "html.parser")
q=soup1.findAll("div", {"class":"categoryList"})
q
```

```python
w=q[0].findAll("a")
w
```

# Books are available in thses topics : 

```python
topics=[]
for i in range(len(w)):
    e=w[i].get('href')
    topics.append(e)
    print(topics[i])
```

# Url for each category and to be used for parsing

```python
url=[]
for i in range(len(topics)):
    url.append("https://b-ok.cc/{0}?page=".format(topics[i].replace(" ", "%20")))
    print(url[i])
```

```python
print(len(url))
```

# Url for counting the total no of pages in each category

```python
url1=[]
for i in range(len(topics)):
    url1.append("https://b-ok.cc/{0}".format(topics[i].replace(" ", "%20")))
    print(url1[i])
```

# Counting the total no. of pages in each category

```python
number_of_pages=[]
for i in range(len(url1)):
    my_url = url1[i]
    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()

    page_soup = soup(page_html, "html.parser")
    links = page_soup.findAll("script")
    data = str(links[-5])
    d=data.replace(' ','').replace('\n', ' ').replace(",", " ")
    d=d[38:40]
    number_of_pages.append(d)
    print(d)
    
print(number_of_pages)
```

```python
print(len(number_of_pages))
```
